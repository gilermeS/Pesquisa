```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as rc
import pandas as pd
```

```python
alg = 'CRNN'

data = pd.read_csv(f"Results/{alg}_results.csv", index_col=0)

def get_mse(i, h0=67.36):
    mse = (i[0] - h0)**2 + i[1]**2

    return i**2
```

```python
data = np.array(data)
```

```python
data[0:]
```

```python
get_mse(data[0:4, 1])
```

```python
# Create figure size in inches
fig, ax = plt.subplots(figsize = (10., 7.))

plot_title=f'{alg}'


# Define axes
ax.set_ylabel(r"$\mathrm{BVT}$", fontsize=32)
ax.set_xlabel(r"$N_z$", fontsize=32)
# ax.set_xlim(17., 90.)
# ax.set_ylim(5e-2, 5e2)
plt.yscale('log')
for t in ax.get_xticklabels(): t.set_fontsize(20)
for t in ax.get_yticklabels(): t.set_fontsize(20)

# plotting BVT results for each sighz/hz and Nz cases
plt.plot([20,30,50,80], get_mse(data[0:4,1]), '->', color='#228B22')
plt.plot([20,30,50,80], get_mse(data[4:8,1]), '-o', color='black')
plt.plot([20,30,50,80], get_mse(data[8:12,1]), '-s', color='red')
plt.plot([20,30,50,80], get_mse(data[12:16,1]), '-*', color='blue')
plt.plot([20,30,50,80], get_mse(data[16:20,1]), '-v', color='magenta')

plt.title(r"{}".format(plot_title), fontsize='27')
plt.legend((r"$\sigma_{H}/H=0.008$", "$\sigma_{H}/H=0.01$","$\sigma_{H}/H=0.03$","$\sigma_{H}/H=0.05$","$\sigma_{H}/H=0.08$"), loc='best', fontsize=12)
plt.savefig(f'Results/{alg}_results_bvt', bbox_inches='tight')
```
