## Feature Importance Analysis

The goal is to determine which redshift (z) values contribute most to the H₀ predictions.

```python
# import numpy as np
from sklearn.inspection import permutation_importance
# from sklearn.metrics import mean_squared_error
# import matplotlib.pyplot as plt

def calculate_redshift_importance(model, x, y, n_renn=10, random_state=42):
    """
    Calculate feature importance based on permutation importance.
    
    For each redshift point z_i, we shuffle its values and measure
    the increase in prediction error. Higher increase = higher importance.
    
    Parameters:
    -----------
    model : keras.Model
        Trained model
    x : np.ndarray
        Feature array with shape (n_samples, n_points, 2) where 2 = (z, H(z))
    y : np.ndarray
        Target values
    n_renn : int
        Number of times to shuffle each feature
    random_state : int
        Random seed for reproducibility
    
    Returns:
    --------
    dict : Dictionary with 'importances', 'uncertainties', and 'redshifts'
    """
    n_samples, n_points, _ = x.shape
    
    redshifts = x[0, :, 0]  # Get redshift values (assuming same for all samples)
    
    importances = np.zeros(n_points)
    uncertainties = np.zeros(n_points)
    
    y_pred = model.predict(x, verbose=0).flatten()
    baseline_mse = mean_squared_error(y, y_pred)
    
    for i in range(n_points):
        feature_importances = []
        
        for _ in range(n_renn):
            x_shuffled = x.copy()
            np.random.seed(random_state + i * n_renn + _)
            
            # Shuffle only the H(z) values at this point (column index 1)
            # Keep z values fixed, shuffle H(z) to break the relationship
            np.random.shuffle(x_shuffled[:, i, 1])
            
            y_pred_shuffled = model.predict(x_shuffled, verbose=0).flatten()
            shuffled_mse = mean_squared_error(y, y_pred_shuffled)
            
            # Importance = increase in MSE when feature is shuffled
            feature_importances.append(shuffled_mse - baseline_mse)
        
        importances[i] = np.mean(feature_importances)
        uncertainties[i] = np.std(feature_importances)
    
    # Normalize to sum to 1 (relative importance)
    total = np.sum(importances)
    if total > 0:
        importances /= total
        uncertainties /= total
    
    return {
        'importances': importances,
        'uncertainties': uncertainties,
        'redshifts': redshifts,
        'baseline_mse': baseline_mse
    }
```

```python
# Calculate permutation importance
print("Calculating permutation importance (this may take a moment)...")
importance_results = calculate_redshift_importance(
    model, x_test, y_test, 
    n_renn=20, 
    random_state=42
)
print(f"Baseline MSE: {importance_results['baseline_mse']:.6f}")
print(f"Number of redshift points analyzed: {len(importance_results['redshifts'])}")
```

```python
def plot_redshift_importance(importance_results, figsize=(12, 6), save_path=None):
    """
    Plot feature importance across redshift values.
    
    Parameters:
    -----------
    importance_results : dict
        Output from calculate_redshift_importance
    figsize : tuple
        Figure size
    save_path : str, optional
        Path to save figure
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    redshifts = importance_results['redshifts']
    importances = importance_results['importances']
    uncertainties = importance_results['uncertainties']
    
    # Bar plot
    ax1.bar(range(len(importances)), importances, yerr=uncertainties, 
            capsize=2, alpha=0.7, color='steelblue', edgecolor='black')
    ax1.set_xlabel('Índice do ponto redshift (z)', fontsize=11)
    ax1.set_ylabel('Importância relativa', fontsize=11)
    ax1.set_title('Importância das features por índice', fontsize=12)
    ax1.grid(True, alpha=0.3, axis='y')
    
    # Plot importance vs redshift value
    ax2.fill_between(redshifts, importances - uncertainties, 
                     importances + uncertainties, alpha=0.3, color='steelblue')
    ax2.plot(redshifts, importances, '-o', markersize=3, 
             linewidth=1.5, color='steelblue', label='Importância')
    ax2.set_xlabel('Redshift (z)', fontsize=11)
    ax2.set_ylabel('Importância relativa', fontsize=11)
    ax2.set_title('Importância vs valor de redshift', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # Mark most important regions
    top_3_idx = np.argsort(importances)[-3:]
    for idx in top_3_idx:
        ax2.annotate(f'z={redshifts[idx]:.2f}', 
                    xy=(redshifts[idx], importances[idx]),
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()
    
    return fig
```

```python
# Plot the feature importance
fig = plot_redshift_importance(importance_results, save_path='../imagens/feature_importance.png')
```
